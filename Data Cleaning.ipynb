```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/DELL/Downloads/Google_Play_Store_Data/Google-Playstore.csv")
df.head()
```

```python
# Treat rows with null values appropriately
# Drop rows missing
df = df.dropna(subset=['Currency', 'App Name', 'Developer Email', 'Size'])


# Fill missing values with meaningful replacements
df['Rating'].fillna(df['Rating'].median(), inplace=True)  # Use median rating for missing ratings
df['Rating Count'].fillna(0, inplace=True)
df['Installs'].fillna(0, inplace=True)
df['Minimum Installs'].fillna(0, inplace=True)
df['Minimum Android'].fillna("Varies with device", inplace=True)
df['Developer Id'].fillna("Unknown Developer", inplace=True)
df['Developer Website'].fillna("Unknown Website", inplace=True)
df['Released'].fillna("Unknown", inplace=True)
df['Privacy Policy'].fillna("No Privacy Policy", inplace=True)

# Final check
print(df.isna().sum())
```

```python
# Check unique values for potential errors or data input errors
print("Unique Content Ratings:", df['Content Rating'].unique())
print("Unique Free Status Value:", df['Free'].unique())
print("Unique Editors Choice Values:", df['Editors Choice'].unique())
print("Unique Ad-Supported Values:", df['Ad Supported'].unique())
print("Unique In-App Purchases Values:", df['In App Purchases'].unique())
```

```python
df.info()
```

```python
#cleaning size column, converting the sizes to have a uniform unit of MB
def convert_size(size):
    if pd.isna(size) or size == "Varies with device":
        return np.nan
    size = size.replace(",", "")  # Remove commas
    if size.endswith('k'):
        return float(size[:-1]) * 0.001  # Convert KB to MB
    elif size.endswith('M'):
        return float(size[:-1])  # Keep MB as is
    elif size.endswith('G'):
        return float(size[:-1]) * 1000  # Convert GB to MB
    else:
        return np.nan  # Handle unexpected values

df['Size'] = df['Size'].apply(convert_size)
```

```python
# Check for duplicates
duplicate_count = df.duplicated().sum()
print(f"Total Duplicate Rows: {duplicate_count}")
```

```python
# Save the cleaned file for further use

df.to_csv('Google-Playstore-data-Cleaned.csv', index = False)
```
